```julia
using TracyWidomBeta,Plots,Distributions,LinearAlgebra
```

```julia
function model_test(β,n)
    F_pdf=TW(β;pdf=true)
    k=10^6
    ei=zeros(k,1)
    for j=1:k
        N=rand(Normal(0,sqrt(2)), n)
        C1=vec(zeros(n-1,1))
        for i=1:n-1
            C1[i]=rand(Chi(β*(n-i)), 1)[1]
        end
        H=(1/sqrt(β))*SymTridiagonal(N,C1)
        ei[j]=n^(1/6)*(eigmax(H)[1]-2*sqrt(n))
    end
    s=Plots.histogram(ei,normed=true,xticks=([-5;-5:1:1;1],[-5;-5:1:1;1]),xtickfontsize=15,ytickfontsize=15,legend=false;bins=-5:0.1:1)
    Plots.plot!(F_pdf,xlims=[-5,1],lw=5,xtickfontsize=15,ytickfontsize=15,legend=false)
    return s
end
```

The following normalized histograms for $n^{1/6}\left(\lambda_{\text{max}}(H^{\beta}_{n})-2\sqrt{n}\right)$ are generated with $10^6$ samples, $n=10^4$. The small positive bias in each histogram can be improved by using larger values of $n$.

For $\beta=3$, we have

```julia
s3=model_test(3,10^4)
```
